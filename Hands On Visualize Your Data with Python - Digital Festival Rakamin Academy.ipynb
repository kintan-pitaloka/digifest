## annotate

$\textbf{Using annotate()}$ documentation [here](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.annotate.html)

It is often useful to annotate a simple plot to provide context. This makes the plot more readable and can highlight specific aspects of the data. Annotations like text and arrows can be used to emphasize specific observations.

Here, you will once again plot enrollment of women in the Physical Sciences and Computer science over time. The legend is set up as before. Additionally, you will mark the inflection point when enrollment of women in Computer Science reached a peak and started declining using plt.annotate().

To enable an arrow, set arrowprops=dict(facecolor='black'). The arrow will point to the location given by xy and the text will appear at the location given by xytext.

```python
# Plot with legend as before
plt.plot(year, computer_sciences, color='red', label='Computer Science') 
plt.plot(year, physical_sciences, color='blue', label='Physical Sciences')
```

```python
plt.figure(figsize=(8,5))
plt.plot(year, computer_sciences, color='red')
plt.ylim(10, 50)

plt.annotate(xy=(1983,37), xytext=(1983+5,37+5), text='maximum',
             fontsize=16,
            arrowprops=dict(
                            arrowstyle="->",
                            connectionstyle="arc3,rad=.3")
            );
```

### Automate find maximum value

```python
# Compute the maximum enrollment of women in Computer Science: cs_max
cs_max = computer_sciences.max()

# Calculate the year in which there was maximum enrollment of women in Computer Science: yr_max
yr_max = year[computer_sciences.idxmax()]

print(cs_max, yr_max)
```

```python
# Plot with legend as before
plt.plot(year, computer_sciences, color='red', label='Computer Science') 
plt.plot(year, physical_sciences, color='blue', label='Physical Sciences')
# plt.legend(loc='lower right')

# Compute the maximum enrollment of women in Computer Science: cs_max
cs_max = computer_sciences.max()

# Calculate the year in which there was maximum enrollment of women in Computer Science: yr_max
yr_max = year[computer_sciences.idxmax()]

# Add a black arrow annotation
plt.annotate('Maximum', xy=(yr_max, cs_max), xytext=(yr_max+5, cs_max+5), 
             arrowprops=dict(
                             arrowstyle="->",
                            connectionstyle="arc3,rad=.3")
             );
```

Let's looping this, let say we want to annotate maximum value on every categories of degrees! How can we do it?

The categories we want to annotate are physcial sciences, computer sciences, & business

```python
fig, ax = plt.subplots(1, 1, figsize = (8,5))
x = year
ys = [physical_sciences, computer_sciences, business]

for y in ys:
    ax.plot(x, y, label=y.name)
    
    cs_max = y.max()
    yr_max = x[y.idxmax()]
    ax.annotate('Maximum', xy=(yr_max, cs_max), xytext=(yr_max+5, cs_max+5), 
                arrowprops=dict(arrowstyle="->",connectionstyle="arc3,rad=.3")
               )
    ax.legend(loc='lower right')
    
ax.set_xlabel('Year')
ax.set_ylabel('Enrollment (%)')
ax.set_title('Undergraduate enrollment of women')
ax.set_ylim(5, 60) # some minor adjustment
plt.savefig('fig/undergraduate_enrollment_of_women.png');
```

## Example 

We want to create illustration like [FiveThirtyEight](https://fivethirtyeight.com/)

We now need to create a graph to help readers visualize the evolution of gender disparity for Bachelors where the situation was really bad for women in 1970. We’ve already set a threshold of 20%, and now we want to graph the evolution for every Bachelor where the percentage of women graduates was less than 20% in 1970. Source for this example is [DataQuest](https://www.dataquest.io/blog/making-538-plots/).

```python
import matplotlib.style as style
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
```

### Take under 20% in 1970 

```python
df = pd.read_csv('csv/percent-bachelors-degrees-women-usa.csv')
women_majors = df.copy()
```

```python
under_20 = women_majors.loc[0, df.loc[0] < 20]
under_20
```

```python
under_20.index
```

### Our first basic plot

```python
under_20_graph = women_majors.plot(x = 'Year', y = under_20.index, figsize = (12,8))
print('Type:', type(under_20_graph))
```

### Change Style and Modify Axis Label

```python
# The previous code
style.use('fivethirtyeight') # use style fivethirtyeight
fte_graph = women_majors.plot(x = 'Year', y = under_20.index, figsize = (12,8))
fte_graph.tick_params(axis = 'both', which = 'major', labelsize = 18) # increase label size

ticks_loc = fte_graph.get_yticks().tolist()
fte_graph.yaxis.set_major_locator(mticker.FixedLocator(ticks_loc))
fte_graph.set_yticklabels(labels = [-10, '0   ', '10   ', '20   ', '30   ', '40   ', '50%', 60]); # modify y label
```

### Add Line in Axis and Signature Bar

```python
# The previous code
fte_graph = women_majors.plot(x = 'Year', y = under_20.index, figsize = (12,8))
fte_graph.tick_params(axis = 'both', which = 'major', labelsize = 18)
fticks_loc = fte_graph.get_yticks().tolist()
fte_graph.yaxis.set_major_locator(mticker.FixedLocator(ticks_loc))
fte_graph.set_yticklabels(labels = [-10, '0   ', '10   ', '20   ', '30   ', '40   ', '50%', 60]);

# Generate a bolded horizontal line at y = 0
fte_graph.axhline(y = 0, color = 'black', linewidth = 1.3, alpha = .7)

# Add an extra vertical line by tweaking the range of the x-axis
fte_graph.set_xlim(left = 1969, right = 2011)

# Remove the label of the x-axis
fte_graph.xaxis.label.set_visible(False)

# The signature bar
fte_graph.text(x = 1965.8, y = -7,
    s = ' ©DATAQUEST Source: National Center for Education Statistics',fontsize = 14, 
               color = '#f0f0f0', backgroundcolor = 'grey')
```

### Add Title and Subtitle

```python
# The previous code
fte_graph = women_majors.plot(x = 'Year', y = under_20.index, figsize = (12,8))
fte_graph.tick_params(axis = 'both', which = 'major', labelsize = 18)
ticks_loc = fte_graph.get_yticks().tolist()
fte_graph.yaxis.set_major_locator(mticker.FixedLocator(ticks_loc))
fte_graph.set_yticklabels(labels = [-10, '0   ', '10   ', '20   ', '30   ', '40   ', '50%', 60]);
fte_graph.axhline(y = 0, color = 'black', linewidth = 1.3, alpha = .7)
fte_graph.set_xlim(left = 1969, right = 2011)
fte_graph.xaxis.label.set_visible(False)
fte_graph.text(x = 1965.8, y = -7,
    s = ' ©DATAQUEST Source: National Center for Education Statistics',fontsize = 14, 
               color = '#f0f0f0', backgroundcolor = 'grey')


# Adding a title and a subtitle
fte_graph.text(x = 1966.65, y = 62.7, s = "The gender gap is transitory - even for extreme cases",
               fontsize = 26, weight = 'bold', alpha = .75)
fte_graph.text(x = 1966.65, y = 57,
               s = 'Percentage of Bachelors conferred to women from 1970 to 2011 in the US for\nextreme cases where the percentage was less than 20% in 1970',
              fontsize = 19, alpha = .85)
```
